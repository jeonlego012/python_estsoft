```python
df = pd.DataFrame({'a': [10,20,30], 'b': [20,30,40]})
df['a']**2
```

```python
# apply 함수 적용
```

```python
def test(x):
    if x > 20:
        return x**2
    else:
        return x**3
```

```python
test(10)
```

```python
df['a'].apply(test)
```

```python
import numpy as np

def avg_2_mod(x, y):
    return (x+y)/2
    
avg_2_mod(df['a'], df['b'])
```

```python
vfunc = np.vectorize(avg_2_mod)
vfunc(df['a'], df['b'])
```

```python
df1 = pd.read_csv('concat_1.csv')
df2 = pd.read_csv('concat_2.csv')
df3 = pd.read_csv('concat_3.csv')

concat_df = pd.concat([df1, df2, df3])
concat_df
```

```python
new_series = pd.Series(['n1','n2','n3','n4'])
pd.concat([concat_df, new_series])
```

```python
pd.concat([concat_df, new_df], ignore_index=True)
```

```python
col_concat = pd.concat([df1, df2, df3, new_df], axis='columns', ignore_index=True)
col_concat
```

```python
from pathlib import Path
files = list((Path('.').glob('concat_*.csv')))
concat_df = pd.DataFrame()
for file in files:
    df = pd.read_csv(file)
    concat_df = pd.concat([concat_df, df], ignore_index=True)
concat_df
```
