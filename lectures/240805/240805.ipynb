#### load dataset

```python
import seaborn as sb
import matplotlib.pyplot as plt
# 기본으로 내장되어있는 디폴트 dataset.
# https://seaborn.pydata.org/examples/anscombes_quartet.html
# 시각화의 중요성을 설명하기 위한
# anscombe's quartet
anscombe = sb.load_dataset('anscombe')
anscombe
```

#### plot(anscombe)

```python
ds1=anscombe[anscombe['dataset']=='I']
print(ds1)
plt.plot(ds1['x'], ds1['y'], 'r--')
plt.show()
```

```python
fig = plt.figure()
axes1= fig.add_subplot(2, 2, 1) #1사분면
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
ds1=anscombe[anscombe['dataset']=='I']
ds2=anscombe[anscombe['dataset']=='II']
ds3=anscombe[anscombe['dataset']=='III']
ds4=anscombe[anscombe['dataset']=='IV']
axes1.plot(ds1['x'], ds1['y'], 'ro')
axes2.plot(ds2['x'], ds2['y'], 'ro')
axes3.plot(ds3['x'], ds3['y'], 'ro')
axes4.plot(ds4['x'], ds4['y'], 'ro')
fig.suptitle('anscombe dataset') # supervisor title
axes1.set_title('dataset I')
axes2.set_title('dataset II')
axes3.set_title('dataset III')
axes4.set_title('dataset IV')
fig.set_tight_layout(True)
plt.show()
```

#### plot(tips)

```python
tips = sb.load_dataset('tips')
tips
```

```python
# histogram, scatter, boxplot
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(data=tips, x='total_bill', bins=10) # histogram
axes1.set_title('tips')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('freq')
plt.show()
```

```python
fig = plt.figure()
reg, ax = plt.subplots()
sb.rugplot(data=tips, x='total_bill', ax=ax) # rug 깔기
sb.histplot(data=tips, x='total_bill', ax=ax)
sb.displot(data=tips, x='total_bill', ax=ax, kde=True, rug=True)
ax.set_title('bill')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip']) # 그래프 산란효과
axes1.set_title('total_bill vs tip')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('tip')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
x=[
    tips[tips['sex'] == 'Female']['tip'],
    tips[tips['sex'] == 'Male']['tip']
]

axes1.boxplot(x=x, labels=['Female', 'Male'])
axes1.set_title('tip of sex')
axes1.set_xlabel('sex')
axes1.set_ylabel('tip')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
colors = {'Female': '#f1a340', 'Male': '#998ec3'}

# scatter plot 산점도
axes1.scatter(data=tips, x='total_bill', y='tip',
              alpha=0.5, # 투명도
              c=tips['sex'].map(colors),
              s=tips['size']**2*10,
             )
axes1.set_title('total bill vs tip')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('tip')
plt.show()
```

```python
scatter, axes1 = plt.subplots()
sb.scatterplot(data=tips, x='total_bill', y='tip', ax=axes1)
axes1.set_title('total bill vs tip')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('tip')
plt.show()
```

```python
sb.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', palette='viridis')
plt.show()
```

```python
# regplot 산점
reg, axes1 = plt.subplots()
sb.regplot(data=tips, x='total_bill', y='tip', ax=axes1)

plt.show()
```

```python
# relplot
facet1 = sb.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='sex',
          kind='scatter', col='day', row='time', palette='viridis', height=1.7)

sb.move_legend(facet1, loc='lower center', bbox_to_anchor=(0.5, 1), ncol=2)

facet1.figure.set_tight_layout(True)
plt.show()
```

```python
# jointplot
joint1 = sb.jointplot(data=tips, x='total_bill', y='tip', kind='resid')
joint1.set_axis_labels(xlabel='total bill', ylabel='tip')
joint1.figure.suptitle('joint total bill, tip', y=1.1)

plt.show()
```

```python
# kdeplot
kde, ax = plt.subplots()
sb.kdeplot(data=tips, x='total_bill', y='tip', ax=ax, fill=True)
plt.show()
```

```python
# countplot
count, ax = plt.subplots()

sb.countplot(data=tips, x='day', ax=ax, palette='YlGnBu')
ax.set_title('days')
plt.show()
```

```python
# barplot
import numpy as np
bar, ax = plt.subplots()
sb.barplot(data=tips, x='time', y='tip', estimator=np.mean, ax=ax)
plt.show()
```

```python
# boxplot
box, ax = plt.subplots()
sb.boxplot(data=tips, x='time', y='total_bill',ax=ax)
plt.show()
```

```python
# violinplot
violin, ax = plt.subplots()
sb.violinplot(data=tips, x='time', y='total_bill', ax=ax)
plt.show()
```

```python
boxviolin, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

sb.boxplot(data=tips, x='time', y='total_bill', ax=ax1)
sb.violinplot(data=tips, x='time', y='total_bill', ax=ax2)

boxviolin.suptitle('box and violin plot')
boxviolin.set_tight_layout(True)
plt.show()
```

```python
# pairplot 관계그래프
sb.pairplot(data=tips)
plt.show()
```

```python
sb.pairplot(tips, hue='sex', palette='viridis')
plt.show()
```

```python
# pairgrid
pg = sb.PairGrid(tips, diag_sharey=False)
pg.map_upper(sb.regplot) # 대각선 기준 위, 아래
pg.map_lower(sb.kdeplot)
pg.map_diag(sb.histplot)
plt.show()
```
