```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# 통계 분석 & 통계 모형 작성 라이브러리
import statsmodels.api as sm
from statsmodels.formula.api import ols

# warning 무시하기
# import warnings
# warnings.filterwarnings('ignore')

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 8

pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 15)
```

```python
red = pd.read_csv('winequality-red.csv', sep=';')
red.head()
```

```python
white = pd.read_csv('winequality-white.csv', sep=';')
white.head()
```

```python
import glob # 폴더를 제어할 수 있게해주는 라이브러리
import os # 경로, 폴더 등을 관리해주는 패키지

all_files = glob.glob(os.path.join('wine*'))
all_files
```

```python
all_data = []
for file in all_files:
    data = pd.read_csv(file, sep=';')
    data['type'] = all_files.index(file)
    all_data.append(data)
all_data
```

```python
wine = pd.concat(all_data, axis=0, ignore_index=True)
wine
```

```python
wine.to_html() # django 템플릿에 보여줄 때
```

```python
red1 = wine.loc[wine.type==0, 'quality']
red1.shape
```

```python
white1 = wine['quality'][wine.type==1]
white1
```

```python
# seaborn
sb.distplot(red1) # 전체분포
sb.distplot(white1)
```

```python
sb.distplot(red1, color="red", vertical=True, label='red wine')
sb.distplot(white1, vertical=True, label='white wine')
plt.legend() # 범례. 기호표시
plt.ylabel('qualityyyyy')
plt.title('quality of red & white')
# matplotlib는 이미지
plt.savefig('quality_of_red_white.pdf', format='pdf')
plt.savefig('quality_of_red_white.jpg', format='jpg')
```

```python
t, p, df = sm.stats.ttest_ind(red, white)
print('t = %.3f, p = %.7f, 자유도 = %d' % (t[0], p[0], df))
```
