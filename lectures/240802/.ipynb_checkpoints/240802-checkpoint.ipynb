```python
# 미션4) 10-100사이의 랜덤한 숫자 10개를 시리즈에 담아서 출력해보기.
# index는 날짜
import numpy as np
import pandas as pd

num = pd.Series(np.arange(10,101))
randomdf1 = pd.Series(num.sample(10).values, index=pd.date_range(start='20240801', end='20240810'))
randomdf2 = pd.Series(num.sample(10).values, index=pd.date_range(start='20240801', end='20240810'))
```

```python
# concat, rename(inplace)
concatdf = pd.concat([randomdf1, randomdf2], axis=1)
concatdf.rename(columns = {0: 'Korea', 1: 'USA'}, inplace=True)
concatdf['avg'] = np.mean(concatdf, axis=1)
concatdf
```
