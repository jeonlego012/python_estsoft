```python
#1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
sales
```

```python
#3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
line, ax = plt.subplots()
sb.lineplot(data=sales, x='month_number', y='facecream', ax=ax, marker='o', color='blue', label='Facecream Sales Data')
sb.lineplot(data=sales, x='month_number', y='facewash', ax=ax, marker='o', color='orange', label='Facewash Sales Data')
sb.lineplot(data=sales, x='month_number', y='toothpaste', ax=ax, marker='o', color='green', label='Toothpaste Sales Data')
sb.lineplot(data=sales, x='month_number', y='bathingsoap', ax=ax, marker='o', color='red', label='Bathingsoap Sales Data')
sb.lineplot(data=sales, x='month_number', y='shampoo', ax=ax, marker='o', color='purple', label='Shampoo Sales Data')
sb.lineplot(data=sales, x='month_number', y='moisturizer', ax=ax, marker='o', color='brown', label='Moisturizer Sales Data')

ax.set_title('Sales Data')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales Units in Number')
plt.xticks(sales['month_number'])
plt.show()
```

```python
#4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
scatter, ax = plt.subplots()
sb.scatterplot(data=sales, x='month_number', y='toothpaste', ax=ax, marker='o', color='blue', label='Toothpaste Sales Data')

ax.set_title('Toothpaste Sales Data')
ax.set_xlabel('Month Number')
ax.set_ylabel('Number of Units Sold')
plt.grid(True, linestyle='--')
plt.xticks(sales['month_number'])
plt.show()
```

```python
#7
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
sales.loc['sum'] = sales.sum()
fig, ax = plt.subplots()
ax.pie([sales.loc['sum','facecream'], sales.loc['sum','facewash'], sales.loc['sum','toothpaste'], sales.loc['sum','bathingsoap'], sales.loc['sum','shampoo'], sales.loc['sum','moisturizer']], labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right', bbox_to_anchor=(1.5, 0))
plt.show()
```

```python
#8
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
line, ax = plt.subplots(nrows=2)
facet0 = sb.lineplot(data=sales, x='month_number', y='bathingsoap', ax=ax[0], marker='o', color='black')
facet1 = sb.lineplot(data=sales, x='month_number', y='facewash', ax=ax[1], marker='o', color='red')

ax[0].set_title('Sales Data of a Bathingsoap')
ax[1].set_title('Sales Data of a Facewash')
ax[0].set_xlabel('Month Number')
ax[0].set_ylabel('Sales Units in Number')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Sales Units in Number')

facet0.figure.set_tight_layout(True)
facet1.figure.set_tight_layout(True)

ax[0].set_xticks(sales['month_number'])
ax[1].set_xticks(sales['month_number'])
plt.show()
```

```python
#9
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

sales = pd.read_csv('company_sales_data.csv')
faceCremSalesData = sales['facecream'].tolist()
faceWashSalesData   = sales['facewash'].tolist()
toothPasteSalesData = sales['toothpaste'].tolist()
bathingsoapSalesData = sales['bathingsoap'].tolist()
shampooSalesData = sales['shampoo'].tolist()
moisturizerSalesData = sales['moisturizer'].tolist()

fig, ax = plt.subplots()
plt.stackplot(sales.month_number, 
              faceCremSalesData, faceWashSalesData, toothPasteSalesData, bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
             labels=['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer'],
             colors=['m', 'c', 'r', 'k', 'g', 'y'],
             )
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales Units in Number')
ax.set_title('All Product Sales Data using stackplot')
plt.legend(loc='upper left')
plt.show()
```
